```python
# https://www.youtube.com/watch?v=FdqDgoG-SFM
# https://python-visualization.github.io/folium/quickstart.html
# pip show pandas
# Import libraries
import folium as fo
import pandas as pd
import streamlit as st
```

```python
#Create map

def mymap():
    items = pd.read_excel('map_list.xlsx')
    world = fo.Map(
        zoom_start=2,
        location=[30, 20])

    # Add markers to the map
    for _, item in items.iterrows():
        fo.Marker(location=(item['latitude'], item['longitude']), 
                  popup=(item['location']), 
                  tooltip=(item['location']), 
                  icon = fo.Icon(color= 'darkred')).add_to(world)
    display(world)
```

```python
mymap()
```
